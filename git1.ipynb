```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from tkinter import filedialog
from tkinter import Tk
```

```python
def upload_image():
    root = Tk()
    root.withdraw()  # Hide the root window
    file_path = filedialog.askopenfilename()  # Open file dialog
    return file_path
```

```python
file_path = upload_image()
image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
```

```python
kernel_size = (3, 3)  # You can change this to other sizes, like (5, 5) or (7, 7)

# Apply the box filter using cv2.blur (which is essentially a box filter)
box_filtered_image = cv2.blur(image, kernel_size)
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('O Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Bo Image')
plt.imshow(box_filtered_image, cmap='gray')
plt.axis('off')

plt.show()
```
